* Using **slices** to create arrays from **other** ndarrays **doesn't** create  **copies**
* To have **distinct** arrays, we have to use the **method copy**


````
 # creating part ndarray from a slice of c ndarray
part = c[:2,0:2]
print("part=\n",part)
part[0,1]=1000
# modification of cell(0,1) of part will 
# alter c values too.
print("now part = \n",part)
print("and c now =\n",c)
````

````
c = np.array([[1,2,3],[4,5,6]])
 # part2 is a slice copy of c
part2 = c[:2,0:2].copy()
print("part2=\n",part2)
part2[0,1]=1000
# modification of cell(0,1) of part2
# will not alter c values.
print("now part2 = \n",part)
print("and c always =\n",c)
````

# 3- Operations with ndarray

## Artithmetic Operations and Linear Algebra

````
# element wise multiplication: cell by corresponding cell,
# matrices with the same shape
res = c * np.eye(3)[0:2,0:3]

# the result is a matrix with the same shape
print ("res=",res)

# matrix multiplication:  line by columuns
# different shapes but:
# number of columus of the first matrix == number of lines of the second matrix
res = c @ np.eye(3)  # same as np.dot(c,np.eye(3)) or c.dot(np.eye(3))

# the result is a matrix wiht:
# number of lines == number of lines of the first matrix
# number of columns == number of columnus of the first matrix
print("res=",res)
````

## Element wise multiplication vs Matrix multiplication

* To perform an **element wise multiplication**, the matrices mus have the **same shape**:

>$c_{2,3} =
 \begin{pmatrix}        
  1 & 2 & 3 \\
 5 & 6 & 7 
 \end{pmatrix}$
 ,  $d_{2,3} =
 \begin{pmatrix}
  1 & -1 &  0\\
 -5 &  1 &  2
 \end{pmatrix}$
 
 
 * **d** has to be **reshaped** to apply **matrix multiplication** with **c**
 
 * The reshape method called from **d** will not alter **d**, but It will create another reference (a view of) to **d**
 
 * After calling reshape with the follwing instruction:
 

```
d=d.reshape(3,2)
```
 **d** will be equal to:      $d_{3,2} =
 \begin{pmatrix}
  1 & -1\\
 0 & -5 \\
 1 & 2
 \end{pmatrix}$
 
 
 

 
 

````
# c * d
d = np.array([[1,-1,0],[-5,1,2]])
res = c * d
print ("res=\n",res)
# c @ d
# reshape method transform the shape of the array
# the number of scalar values remain the same
d=d.reshape(3,2)
print("new d=\n",d)
res = c.dot(d)
print("res2=\n",res)
````

## Linear Algebra

* We already seen the matrix multiplication using **dot** **method**
or **np.dot function** or the operator **@**

* There are other functions related to linear Algebra as: **diag**,**trace**,
**inv**, **solve**, … etc.

* as for sacalrs, matrices have inverse regarding the matrix multiplication operation:
  
  >$mat * mat^{-1}= I$
  
  $I$  is the Identity matrix
  
* A system of linear equations can be represented by matrices:

  >$ mat * x = y $
  
  so: 
  
  >$ x=  mat^{-1} * y$
  
  
 * for example: 

  >$   x_{1} +  x_{2} = 4 $
  
  >$   x_{1} -  x_{2} = 0 $
  
  
  * in this case: 
  >$mat_{(2,2)} =
 \begin{pmatrix}        
  1 & 1  \\
 1 & -1 
 \end{pmatrix}$
 ,  $x_{(2,1)} =
 \begin{pmatrix}
  x_{1} \\
 x_{2}
 \end{pmatrix}$
 
 >   $y_{(2,1)} =
 \begin{pmatrix}
  4\\
 0
 \end{pmatrix}$
  
  and solution will be:
  
 $x_{(2,1)} =
 \begin{pmatrix}
  2 \\
 2
 \end{pmatrix}$

    

  


````
# diag returns the diagonal of a square matrix
mat = np.random.randn(3,3)
print ("mat==",mat)
print("mat diagonal ==", np.diag(mat))
# trace retruns the sum of the diagonal elements
print(np.trace(np.eye(4)))
# inv return the inverse of a square matrix : mat * inv(mat)== identity matrix
print("mat inverse==",np.linalg.inv(mat))
# solve return the solution of the equation Ax=B (the values of x)
print("solution of mat * x = I is: ",np.linalg.solve(mat,np.eye(3,3)))
````

# 4- File loading and Saving

* It is possible to **save** and **load**  ndarrays into **binray format**

````
# save c to "file_c.npy"
np.save("file_c",c)
# loading c from "file_c.npy" into c2
c2= np.load("file_c.npy")
print("c2==",c2)
# saving multiple ndarrays: c and d into "files.npz"
np.savez("files",c=c,d=d)
#loading c and d from "files.npz"
res = np.load("files.npz")
print("c==",res["c"])
print("d==",res["d"])
````

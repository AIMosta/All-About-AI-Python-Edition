# 1-Decision Trees

## concept

* Decision Trees  are an other type of **learning models** . **Trained** on **labeled** data.


* They are used for both **classification** and **regression**.

* The model is **trained** by  **spliting** the dataset in order to** reduce**  its **impurity** using a **tree structure**.

* The **non-leaf** nodes are the  **decision nodes**:  They perform a **test **  or **apply** a **rule** on feature values. The result of the test will **split** the data into other sets ==> other nodes

* The** leaf nodes:** are **non splittable** nodes. They represent the **final result** of the classification or the regression.



## Binary trees

* each **node** will split the data into **2 sets**.

## Decision Trees in scikit-learn


* Scikit-learn implement an optimized version of the Classification And Regression Tree (CART) algorithm.

* It is a Binary tree that uses **one feature**  to be tested regarding ** one threshold** to split the data.

* The chosen feature and  threshold are those which **minimize the impurity value** after each split <==> **information gain**

* The impurity caluclation formula, can be considered as the **cost function**. 

* For classification, scikit-learn implements the following **impurity measures**: **gini**, **entropy** and** miclassification** measures.

* For regression, it implements these measures: the **mean squared** and **mean absolute errors**.




# 2- Impurity calculation

* All these formulas apply to the algorithm used in **scikit-learn** (binary decision tree).
* It is computed from the number of different classes existing in the dataset related to a node.
* the size of the dataset related to the  **root** node is the size of training dataset. 

*  For classfication, the following term is used in both **entropy** and **gini** measures:

> > $p_{mk} = 1/ N_m * \sum_{x_i \in R_m} I(y_i = k)$
* $m$ corresponds to the node **m**

  * $k$ corresponds to the class $k$.
  * $ N_m$ size of dataset corresponding to the node $m$ == the number of all the observations in node $m$.
  * the formula means:  proportion of class $k$ observations in node $m$.

*  For a regresesion, the following term is used in both **mean squared** and **mean absolute errors** :




\begin{align}\begin{aligned}\bar{y}_m = \frac{1}{N_m} \sum_{i \in N_m} y_i\end{aligned}\end{align}



## Entropy

*  $X_m$ is the training data in node $m$.

* **Entropy value for a node**

$H(X_m) = - \sum_k p_{mk} \log_2(p_{mk})$

with the following convention: $0* log_20 = 0$.

* **Entropy value at a level **

$J= \frac {N_{left}}{(N_{left} +N_{righ}) } * H(X_{left.node}) +\frac {N_{right}}{(N_{left} +N_{righ}) } * H(X_{right.node})  $



## Gini
* where $X_m$ is the training data in node $m$.
* **Entropy value for a node**

  $H(X_m) = \sum_k p_{mk} (1 - p_{mk}) = 1 -\sum_k p_{mk}^2 $

* **Entropy value for a level:**

    Same  formula as in the previous one.

## Mean and absolute squred errors

* **mean squared error at a node $m$:**

\begin{align}\begin{aligned}\\H(X_m) = \frac{1}{N_m} \sum_{i \in N_m} (y_i - \bar{y}_m)^2\end{aligned}\end{align}

* **mean absolute error at a node $m$:**

\begin{align}\begin{aligned}\\H(X_m) = \frac{1}{N_m} \sum_{i \in N_m} |y_i - \bar{y}_m|\end{aligned}\end{align}


*  **impurty value at a level of the tree**:

      calculated the same way as in classification.

## Thresholds and features selection: using entropy

## Training

* After the selection of the (feature, threshold ) pair, the root node is split into 2 other  nodes using that pair.

* The process is repeated for the new nodes until:

 * Max depth of the tree is reached (the number of levels ) or

 * Min samples by node is reached or

 * Other parameters or

 * No other split that reduces impurity is possible.

* A value is attributed to each final node (the leaf):
 * For a classification: it is a class number that represents the most represented class in that node.
 * For a regression: it is the average values of the y labels corresponding to all the samples belonging to that node.

## Predicting

* You start from the root: you compare the values of the features selected in that node with the threshold value of the noce. 
* The result : if it is True you go to the left, else you go to the right node.
* You repeat this process, until you rich a leaf.

* Classification
 * If you are doing classification, the predicted class for your sample, will be the class corresponding to that leaf.
  * A probability of belonging to that class can be computed: count of samples belonging to that class (in that node) / count of all samples (in that node)
* Regression
 * If your are doing a regression, the predicted value will be the value related to that leaf.

# 4- Example: a decision tree classifier

##  Training

```
%matplotlib inline

import matplotlib.pyplot as plt


# import numpy
import numpy as np
#import train_test_split from model_selection
from sklearn.model_selection import train_test_split
```

```


#import iris plant datasets tools
from sklearn.datasets import load_iris
# import the decision tree classifier
from sklearn.tree import DecisionTreeClassifier

myIris = load_iris()
X = myIris.data
y = myIris.target
x_train,x_test,y_train,y_test= train_test_split(X, y, test_size=0.25)
myModel = DecisionTreeClassifier(max_depth=3)
myModel.fit(x_train, y_train)
```

```
print(myIris.DESCR)
```

## Predicting

* In this section we will test our classifier using the test sets. We will see the results as:
  * the list of predicted classes
  * the list of all the probabilities claculated probabilities in each node (the portion of traing samples belonging to a specefic class in each node).
  * print the decision path:  (i,j) pairs, for each sample i, it returns all the nodes j it vsited.
  
 


```
y_pred = myModel.predict(x_test)
y_pred
```

```
y_pred_prob= myModel.predict_proba(x_test)
y_pred_prob
```

```
print(myModel.decision_path(x_test))
```

```
print(x_test[0])
```

## Analyze the results

* We will measure the preformances by the confusion matrix and the classification report.

* from the confusion matrix, we see that the classes 0 and 2 were all well predicted. For the class 1 , 

```
# import the confusion matrix
from sklearn.metrics import confusion_matrix
myConfMat = confusion_matrix(y_test,y_pred)
print( myConfMat)
```

# 5- Example: a decision tree regressor

## Training

* For the data, we will generate the random points using **sklearn.datasets.make_reggression**

```

from sklearn.datasets import make_regression
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

# generate a random regression problem
xr,yr= make_regression( n_features=2)



fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xr[:,0], xr[:,1], yr,marker="^",c="r")
ax.set_xlabel('feature 1')
ax.set_ylabel('feature 2')
ax.set_zlabel("Labels")

plt.show()

```

```
from sklearn.tree import DecisionTreeRegressor

xr_train,xr_test,yr_train,yr_test= train_test_split(xr, yr, test_size=0.25)

myModelR = DecisionTreeRegressor(max_depth=6)
myModelR.fit(xr_train, yr_train)
```

## Predicting:

* To test our model, we will:
  * draw the generated model byt predicting the y labels  for a continous new x values.
  * test the model with the testing set.
  

```

%matplotlib inline  


xr_new_f1= np.linspace(-3,3,1000)
xr_new_f2= np.linspace(-3,3,1000)
xr_new1 = xr_new_f1.reshape(-1,1)
xr_new2 = xr_new_f2.reshape(-1,1)
xr_new = np.append(xr_new1,xr_new2,axis=1)


fig=plt.figure()

yrPred1 = myModelR.predict(xr_new)

ax2 = fig.add_subplot(111, projection='3d')
ax2.scatter(xr_train[:,0],xr_train[:,1],yr_train)
ax2.set_xlabel('feature 1')
ax2.set_ylabel('feature 2')
ax2.set_zlabel("Labels")


ax3 = fig.gca(projection='3d')

ax3.plot(xr_new_f1,xr_new_f2, yrPred1,c="r")

plt.tight_layout()

plt.show()



```

```
yrPred2 = myModelR.predict(xr_test)
fig=plt.figure()
ax4 = fig.add_subplot(111, projection='3d')
ax4.scatter(xr_test[:,0],xr_test[:,1],yr_test,c="g",marker=".")
ax4.set_xlabel('feature 1')
ax4.set_ylabel('feature 2')
ax4.set_zlabel("Labels")
ax4.scatter(xr_test[:,0],xr_test[:,1],yrPred2,c="r",marker="+")
plt.tight_layout()
plt.show()
```

## Analyse the resuts

* We will use the **score** that returns the value of **R2 coefficient of determination**. 
* the parameters of the moethod are: $x_{test}$ : the test samples, and $y_{true}$ the true labels.
* It is calculated as follow:
$R²=1-  \frac{u}{v} $
> where

    $u =  \sum(y_{true} - y_{pred}) ^2 $
    
>>and

>$v=\sum(y_{true }- \hat y_{true})^2$
>>where
$\hat y_{true}$ is the mean of $y_{true}$


* The best possible score is **1.0**.


```
# the r2 coefficient determination
myModelR.score(xr_test,yr_test)
```

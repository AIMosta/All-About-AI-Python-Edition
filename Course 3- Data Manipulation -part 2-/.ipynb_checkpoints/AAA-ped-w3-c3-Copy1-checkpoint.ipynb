# 1- Date and time types

## Datetime module

```python
# importing datetime module,
import datetime as dt
```

```python
# datetime type
# to have the current time
now =dt.datetime.now()
now
```

```python
# to access the year 
now.year
```

```python
# datetime.timedelta type
diff = dt.datetime.now() - dt.datetime(2018,10,26)
diff
```

```python
#  difference in days
diff.days
```

```python
# timedelta type
# operations with timedelta
dt.datetime.now()+dt.timedelta(2)*2
```

```python
# string conversions
dateS = dt.datetime(2018,11,5)
# from datetime to string using str
str(dateS)

```

```python
# from datetime to string using strftime
dt.datetime.now().strftime("%Y/%m/%d, %H:%M")
```

```python
# from string to datetime 
from dateutil.parser import parse
parse("2018/10/27, 13h05")
```

# 3- Date ranges, frequencies and shifting

## Date ranges

```python
# creating dateitmeindex using a date range
dr = pd.date_range("2018-10-05","2018-10-10")
dr
```

```python
# creating the same datetimeindex differently 
# with start and period key argument
dr = pd.date_range(start= "2018-10-05",periods= 6)
dr
```

```python
# creating the same datetimeindex differently 
# with start and period key argument
dr = pd.date_range(end="2018-10-10",periods= 6)
dr
```

## Frequencies (using date_range)

```python
# creating a datetimeindex  with an hour frequency
dr2 = pd.date_range(start="2018-10-05",periods =6,freq="H")
dr2
```

```python
# creating a datetimeindex  with 3 hours frequency
dr3 = pd.date_range(start="2018-10-05-15-00",periods =6,freq="3h")
dr3
```

```python
# creating a datetimeindex  with 30 minutes frequency
dr4 = pd.date_range(start="2018-10-05-15-00",periods =6,freq="30min")
dr4
```

```python
# creating a datetimeindex  with month begin frequency
dr5 = pd.date_range(start="2018-10-05",periods =6,freq="MS")
dr5
```

## Frequencies (using rsample)

```python
# resample the series by a day frequency
ser1.resample("d").mean()
```

```python
# resample the series by a 12 hours frequency
ser2.resample("12h").mean()
```

## Shifting


```python
# shifting values without shifting indexes
ser1.shift(1)
```

```python
# shifting values with indexes
ser1.shift(1,freq="D")
```

```python
from pandas.tseries.offsets import Day, Week
from datetime import date

#shift using time offsets
now = dt.datetime.now()

print(now.strftime("Today = %Y/%m/%d"))
# timestamp created by shifting by 1 day from now
tomorrow = now +Day()
print(tomorrow.strftime("Tomorrow = %Y/%m/%d"))
```

```python
# shift to the next week using rollforward
Week().rollforward(now).strftime("Next week = %Y/%m/%d")
```

```python
# shift to the previous week using rollback
Week().rollback(now).strftime("Previous week = %Y/%m/%d")
```

# 4- Timezone

## Timezone conversions

```python
# from a naive (without time zone) serieses to a localized  one
print("ser1 timezone:",ser1.index.tz)
ser3=ser1.tz_localize("Africa/Algiers")
ser3.index.tz
```

```python
ser1
```

```python
ser3
```

```python
# from a localized serieses to another time zone series
print("ser3 timezone:",ser3.index.tz)
ser4= ser3.tz_convert("America/New_York")
ser4.index.tz
```

```python
ser4
```

```python
# generating a localized date_range
ser5 = S(np.random.rand(1),index= pd.date_range(start= "2018-10-27",periods=1,tz="Africa/Algiers"))
ser5
```

```python
# additionning 2 series with different time zones
# the result will be in UTC time zone 
(ser3 + ser4).index.tz
```

```python
# additionning 2 series with different time zones
ser3 +ser4 
```

# 5- Periods and resampling

## Periods

```python
# Periods represent durations in time. 
# aPer will represent a duration equal to a month
aPer = pd.Period("January 2018",freq="M")
aPer
```

```python
# adding values to a pariod will shift the period to that value * frequency
aPer + 5
```

```python
# period_range will create a range of periods and also a PeriodIndex
aPerR = pd.period_range("January 2018","December 2018",freq="M")
aPerR
```

```python
# period indexes can be used as Series indexes
# (they can also be creatted using pd.PeriodIndex)
ser6 = S(range(12),index=aPerR)
ser6
```

## Periods conversions

```python
# we can convert periods from frequency to another
aPerR.asfreq("D",how="start")
```

```python
ser1
```

```python
# time series indexed by timestamps (datetime indexes) can be
# converted to series indexed by periodindexes
ser1.to_period("M")
```

```python
ser6
```

```python
#converting to timestamp indexes values
ser6.to_timestamp()
```

## Resampling

```python
ser6
```

```python
# resampling can be done in 2 directions 
# from hight to low frequency = downsampling
# ( from shorter to longer time duration)
ser6.resample("A-DEC").sum()
```

```python
# from low to high frequency = upsampling
# ( from longer to shorter time duration) 
ser6.resample("D").asfreq()
```

```python
ser6.resample("D").asfreq().dropna()
```

# 6- Rolling and expanding

## Rolling

```python
# creating ser7 series using a period index
ser7 = S(range(10),index=pd.period_range("20180101",periods=10))
ser7
```

```python
# rolling.mean() (in this case)will calculate the mean
# of each 2 consecutive values
ser7.rolling(2).mean()
```

```python
# to avoid generaing nan values (like in this case)
# or to take into account the fact that some periods data can be missing
# we can use "min_periods" key argument
ser7.rolling(2,min_periods=1).mean()
```

```python
# creating and plotting ser8
ser8= S(np.random.lognormal(size=500),index= pd.period_range("20180101",periods=500))
ser8.plot()
```

```python
# rolling.mean in this case will calculate
# the mean of each 50 consecutive values
ser8.rolling(50).mean().plot()
```

## Expanding

```python
#  expanding.mean (in this case )wil caclulate 
# the mean of 1, then 2, then 3, ..., until 10 consecutive values
ser7.expanding().mean()
```

```python
# apply expanding.mean to the ser8 series
ser8.expanding(50).mean().plot()
```

```python
# necessary imports
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

## Usage

```python
# you can specify your own country center lattitude and longitude coordinates (in decimal degrees)
country_lat=28.0339
country_long=1.6596
# you can specify longitude and latitude for your own city (in decimal degrees)
city_lon =0.1401 
city_lat = 36.0131

# a figure with a defined size
plt.figure(figsize=(8, 8))
# selecting orthographic projection
m = Basemap(projection='ortho',lat_0=country_lat, lon_0=country_long)
# display the blumarble image as map background  
m.bluemarble();
# convert the geographic coordinates to projection coordinates
x, y = m(city_lon,  city_lat)
#plotting the mark corresponding to the city coordinates
plt.plot(x, y, 'ok', markersize=5)
# plotting a text showing the name of the city
plt.text(x+300000, y,"Mostaganem", bbox=dict(facecolor="green"),fontsize=12,color="black");
# will draw political country bondaries
m.drawcountries()
```

```python
# download the needed File
# the csv file, is downloaded from: https://simplemaps.com/data/dz-cities
# you can download other files corresponding to your country from 
# the same website and use them


!git clone https://github.com/MostaSchoolOfAI/AAA-Ped-Week3.git
 
```

```python
cities = pd.read_csv('AAA-Ped-Week3/A3P-w3-dz.csv')
# display only the first row
cities.head(1)
```

## Plotting

```python
# extract latitues and longitues values from corresponding columns
lat = cities['lat'].values
lon = cities['lng'].values
# extract the population values
population = cities['population'].values

fig = plt.figure(figsize=(8, 8))
# use of , you have to specify the width and the height( in projections units : meters)
# or specify the four corner coordinates
m = Basemap(projection='lcc',lat_0=country_lat, lon_0=country_long,width=5E6,height=5E6)
# Project a shaded relief image onto the map
m.shadedrelief()
m.drawcountries(color='black')
# scatter city data, with size reflecting population
m.scatter(lon, lat, latlon=True,s= population/10000,c="red",alpha=0.5 )

# selecting 3 values: 10000, mean, maximum of population values as legend
legend_values= [10000,int(cities.population.mean()),int(cities.population.max())]

# plotting the legend
for a in legend_values:
 plt.scatter([], [], c='k', alpha=0.5, s=a/10000,label=str(a) )

plt.legend(loc='upper left');
```

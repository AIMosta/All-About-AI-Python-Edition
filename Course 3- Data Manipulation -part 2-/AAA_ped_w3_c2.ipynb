# Data manipulation:  Plotting and visualization

# 1- matplotlib

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series as S, DataFrame as DF
```

```
%matplotlib inline
```

```
# creating a Series
ser1 = S(range(3,10),index=list("abcdefg"))

ser1
```

```
# plotting ser1 values as a line
plt.plot(ser1)
```

## Basic plotting

```
# annother way to plot lines (figures)
# using figures, axis and subplots
# creating a figure instance
myFig= plt.figure()
# adding to the figure a subplot:
# divinding the corresponding region
# in (1,1) ==> 1 line x 2 columns
# the third value 1: position (number of the subplot)
axis1= myFig.add_subplot(1,2,1)
axis2= myFig.add_subplot(1,2,2)
# plotting in the first position
axis1.plot(ser1)
```

## Customizing plots

```

# you can specify : the color, the line_style
# and the marker type of the plot
plt.plot(ser1, color='g', linestyle='dashed', marker='o')
```

```
myFig2= plt.figure()
axis1= myFig2.add_subplot(1,1,1)
# modify the values of x to be shown
axis1.set_xticks([0,2,4,6])
# modify theire names
axis1.set_xticklabels(list("ABCD"))
# adding a title to the plot
axis1.set_title ("Plotting ser1")
# ploting axis1 wiht a label "range"
axis1.plot(ser1,label="range",marker="o")
# adding a legend 
axis1.legend(loc="best")
```

## Annotations

```
# adding a text in a plot at a certain position
plt.text(0.5,0.5,"Here is the example")
```

```
# adding a figure in  a plot at a certain position
rec= plt.Rectangle((0.5, 0.5), 0.2, 0.2, color='r')
axis = plt.figure().add_subplot(1,1,1)
axis.add_patch(rec)
```

```

axis = plt.figure().add_subplot(1,1,1)
# add a text anotations+ arrow :
# the arrow will be added if th xytext position
# is different from xy position
axis.annotate("My red annotation", xy=(2,5),
xytext=(2,5-2),              
arrowprops=dict(facecolor='red', width=3),horizontalalignment='right')
# add an other text anotations+ arrow :
# the y position of the text is higher than xy annotation
# so by default the arrow will be drawn under the text
axis.annotate("My green annotation", xy=(4,7),
xytext=(4,7+2),              
arrowprops=dict(facecolor='green', width=3),horizontalalignment="left")
  
axis.plot(ser1)
```

* You can **customize** the **default** options of matplotlib plots, you can use the **rc**  method

```
# set the default size of a figure 
#  to 3 x 3
plt.rc('figure', figsize=(3, 3))
# a dict of font options
font_options = {'family' : "Courier New, monospace",
'weight' : "bold",'size':2.5}
# customizing the default font 
plt.rc('font', **font_options)
plt.plot()
```

## Histograms

```
# creating a dataframe with random values
df3 = DF(np.random.rand(40,3),columns=list("ABC"))
```

```
# plotting the histogram corresponding to the number 
# of values of column "C" in each interval (bin) from 50 intervals
# having all the same length
df3.loc[:,"C"].hist(bins=50)
```

```
# plotting the histograms of all columns
df3.hist(bins=50)
```

## Points (scatter) plot with seaborn

```
# plotting x,y points (values from B, C columns 
#respectively) with a linear regression fit
sns.regplot("B","C",data=df3)
```

```
# plotting all the columns by pairs
# the diagonal represents the corresponding
# histogram for each column 
sns.pairplot(df3)
```

## Categorical Data

```
# creating a dataframe with values tha can be grouped 
# in 2 dimensional grouping
df4 = DF ({"col1":list(range(5,8))*4,"col2":np.random.rand(12),"col3":[1]*3 + [2]* 3 +[1]*3+[2]*3,"col4":["cat1"]*6 +["cat2"]*6})
df4
```

```
# ploting the values as bars regarding 
# two groupings : col3 values,and col4 values
sns.factorplot(x="col1",y="col2",hue="col3",col="col4",data=df4,kind="bar")
```
